```python
data=open("/content/drive/MyDrive/Data Mining/HW3-1_iris_new_data.txt","r")
data=data.readlines()
print(data)
```

```python
d = []
for l in data:
    s=str(l[0:len(l)-1])
    d.append(s)
print(d)
```

```python
test = []

k = 0
for i in d:
  y = i.split()
  test.append(y)
print(test)

#print(test[0][1])
```

```python
newList = []
listSize = len(test)
for i in range(0, listSize):
    newList.append([])
    for j in range(0, 4):
        newList[i].append(float(test[i][j]))
print(newList)
```

```python
centroid1 = [1,1,1,1]
centroid2 = [2,2,2,2]
centroid3 = [3,3,3,3]
```

```python
i = 0
res1 = []
while i < 150:
  r = abs(centroid1[0] - newList[i][0]) + abs(centroid1[1] - newList[i][1]) + abs(centroid1[2] - newList[i][2]) + abs(centroid1[3] - newList[i][3])
  i += 1
  res1.append(r)

print(res1)
```

```python
i = 0
res2 = []
while i < 150:
  r = abs(centroid2[0] - newList[i][0]) + abs(centroid2[1] - newList[i][1]) + abs(centroid2[2] - newList[i][2]) + abs(centroid2[3] - newList[i][3])
  i += 1
  res2.append(r)

print(res2)
```

```python
i = 0
res3 = []
while i < 150:
  r = abs(centroid3[0] - newList[i][0]) + abs(centroid3[1] - newList[i][1]) + abs(centroid3[2] - newList[i][2]) + abs(centroid3[3] - newList[i][3])
  i += 1
  res3.append(r)

print(res3)
```

```python
clus = []
i = 0

while i < 150:
  m = min(res1[i],res2[i],res3[i])
  if m == res1[i]:
    val = 1
  if m == res2[i]:
    val = 2
  if m == res3[i]:
    val = 3
  clus.append(val)

print(clus)
```
